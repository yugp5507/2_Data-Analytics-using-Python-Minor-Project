```python
df.isna().sum()
```

```python
df['SGPA'] = df['SGPA'].replace('--', 0) 
df['SGPA'] = pd.to_numeric(df['SGPA'], errors='coerce').fillna(0)
df.to_csv(cleaned_file_path, index=False)
```

```python
df['Total Marks'] = df['Total Marks'].astype(str).str.split('/').str[0].str.strip()
df['Total Marks'] = pd.to_numeric(df['Total Marks'], errors='coerce').fillna(0)
print(df['Total Marks'].head())
df['Total Marks'].dtype
df.to_csv(cleaned_file_path, index=False)
```

```python
marks_columns = [
    
    # JAVA
    'JAVA_Ext', 'JAVA_Int', 'JAVA_Total',
    'JAVA_Th_Ext', 'JAVA_Th_Int',
    'JAVA_Pr_Ext', 'JAVA_Pr_Int',

    # .NET
    'NET_Ext', 'NET_Int', 'NET_Total',
    'NET_Th_Ext', 'NET_Th_Int',
    'NET_Pr_Ext', 'NET_Pr_Int',

    # MAD
    'MAD_Ext', 'MAD_Int', 'MAD_Total',
    'MAD_Th_Ext', 'MAD_Th_Int',
    'MAD_Pr_Ext', 'MAD_Pr_Int',

    # IOT
    'IOT_Ext', 'IOT_Int', 'IOT_Total',

    # OSS
    'OSS_Ext', 'OSS_Int', 'OSS_Total',

    # WEB DESIGN
    'CC_Ext', 'CC_Int', 'CC_Total',

    # BMP
    'BMP_Ext', 'BMP_Int', 'BMP_Total'
]

df[marks_columns] = (
    df[marks_columns]
    .apply(pd.to_numeric, errors='coerce')
    .fillna(0)
    .astype(int)
)

df.to_csv(cleaned_file_path, index=False)
```
